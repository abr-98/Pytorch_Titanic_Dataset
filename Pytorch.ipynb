```python
from torch.utils.data import Dataset, DataLoader
```

### Pytorch Dataset Generation and Loading

```python
class Gen_Dataset(Dataset):
    
    def __init__(self,file, mode='train'):
        self.mode=mode
        df=pd.read_csv(file)
        
        if self.mode=='train':
            self.oup=df['Survived'].values.reshape(len(df),1)
            df1=df.drop(['Survived'],axis=1)
            self.inp=df1.values
        else:
            df1=df.drop(['Survived'],axis=1)
            self.inp=df1.values
    
    def __len__(self):
        return len(self.inp)
    
    def __getitem__(self,idx):
        if self.mode=='train':
            inpt=torch.tensor(self.inp[idx])
            oupt=torch.tensor(self.oup[idx])
            
            return { 'inp': inpt,
                    'oup':oupt,
                   }
        else:
            inpt=torch.Tensor(self.inp[idx])
            return { 'inp': inpt
                   }
            
        
        
```

```python
data=Gen_Dataset('cleaned_train.csv')
```

```python
data_train=DataLoader(dataset=data,batch_size=32,shuffle=False)
```

### Modelling

```python
import torch.nn as nn
import torch.nn.functional as func
```

```python
class Neural_Net(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.fc1=nn.Linear(31,64)
        self.fc2=nn.Linear(64,128)
        self.fc3=nn.Linear(128,128)
        self.fc4=nn.Linear(128,256)
        self.fc5=nn.Linear(256,1)
        
    def forward(self,x):
        
        x=func.relu(self.fc1(x))
        x=func.relu(self.fc2(x))
        x=func.relu(self.fc3(x))
        x=func.relu(self.fc4(x))
        x=func.sigmoid(self.fc5(x))
        
        return x
        
```

```python
from torch.optim import Adam
```

```python
criterion=nn.MSELoss()
```

```python
Epochs=200
```

```python
netw=Neural_Net()
```

```python
netw
```

```python
optm=Adam(netw.parameters(),lr=0.001)
```

```python
def train(model,x,y,optimizer,criterion,batch_size):
    model.zero_grad()
    output=model(x.float())
    #print(output)
    #print(y)
    #loss=0
    #for i in range(len(output)):
    loss=criterion(output,y.float())
    #loss=loss/batch_size
    loss.backward()
    optimizer.step()
    
    return loss, output
    
```

### Training

```python
for epoch in range(Epochs):
    epoch_loss = 0
    correct = 0
    for bidx, batch in tqdm(enumerate(data_train)):
        x_train, y_train = batch['inp'], batch['oup']
        #x_train = x_train.view(-1,31)
        #y_train = y_train.view(-1,1)
        #print(x_train)
        #print(y_train)
        if div==1:
            x_train = x_train.to(device)
            y_train = y_train.to(device)
        loss, predictions = train(netw,x_train,y_train, optm, criterion,batch_size=32)
        for idx, i in enumerate(predictions):
            i  = torch.round(i)
            if i == y_train[idx]:
                correct += 1
        acc = (correct/len(data))
        epoch_loss+=loss
    print('Epoch {} Accuracy : {}'.format(epoch+1, acc*100))
    print('Epoch {} Loss : {}'.format((epoch+1),epoch_loss))
```
